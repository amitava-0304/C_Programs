# Decimal to Binary Conversion in C  
## Comparison of Division Method and Bitwise Method

This notebook explains two methods to convert a decimal number into binary:

1. Using division `/` and modulus `%`
2. Using bitwise right shift `>>` and bitwise AND `&`

We will compare:

- Program logic
- Dry run
- Time complexity
- Space complexity
- Which method is faster

# 1. Method 1: Using Division and Modulus

In this method:

- `% 2` gives the binary digit.
- `/ 2` reduces the number.
- Digits are stored from right to left.

Example:

```text
112 / 2 = 56 remainder 0
56 / 2  = 28 remainder 0
28 / 2  = 14 remainder 0
14 / 2  = 7  remainder 0
7 / 2   = 3  remainder 1
3 / 2   = 1  remainder 1
1 / 2   = 0  remainder 1
```

So binary of 112 in 8 bits is:

```text
01110000
```

## C Program Using Division and Modulus

```c
#include <stdio.h>

char *decimalToBinary(unsigned int num, int bits)
{
    static char binary[100];
    int i;

    for(i = bits - 1; i >= 0; i--)
    {
        binary[i] = (num % 2) + '0';
        num = num / 2;
    }

    binary[bits] = '\0';

    return binary;
}

int main()
{
    unsigned int decimal;
    int bits;

    printf("Enter decimal number: ");
    scanf("%u", &decimal);

    printf("Enter number of bits: ");
    scanf("%d", &bits);

    printf("Binary = %s\n", decimalToBinary(decimal, bits));

    return 0;
}
```

# 2. Method 2: Using Bitwise Operators

In this method, we use:

```c
>>
```

and

```c
&
```

Meaning:

- `num >> i` shifts the bits of the number to the right.
- `& 1` checks the last bit.
- If the last bit is `1`, it prints `1`.
- If the last bit is `0`, it prints `0`.

Example:

```c
(num >> i) & 1
```

This extracts the bit at position `i`.

## C Program Using Bitwise Operators

```c
#include <stdio.h>

char *decimalToBinaryBitwise(unsigned int num, int bits)
{
    static char binary[100];
    int i;

    for(i = bits - 1; i >= 0; i--)
    {
        binary[bits - 1 - i] = ((num >> i) & 1) + '0';
    }

    binary[bits] = '\0';

    return binary;
}

int main()
{
    unsigned int decimal;
    int bits;

    printf("Enter decimal number: ");
    scanf("%u", &decimal);

    printf("Enter number of bits: ");
    scanf("%d", &bits);

    printf("Binary = %s\n", decimalToBinaryBitwise(decimal, bits));

    return 0;
}
```

# 3. Time Complexity Comparison

Both methods use one loop that runs according to the number of bits.

For example:

```text
bits = 8
```

Loop runs 8 times.

```text
bits = 16
```

Loop runs 16 times.

```text
bits = 32
```

Loop runs 32 times.

So for both methods:

```text
Time Complexity = O(bits)
```

## Why Time Complexity Is Same

To print an 8-bit binary number, we must generate 8 digits.

Example:

```text
01110000
```

There are 8 digits, so at least 8 operations are required to produce the output.

For `n` bits, we must generate `n` binary digits.

Therefore, the time complexity cannot be less than:

```text
O(bits)
```

So both methods have the same Big-O complexity.

# 4. Which Method Is Faster?

Although both methods have the same Big-O time complexity, the bitwise method is faster in practice.

## Division Method

```c
binary[i] = (num % 2) + '0';
num = num / 2;
```

This uses:

- Modulus `%`
- Division `/`

Division and modulus are comparatively heavier operations.

## Bitwise Method

```c
binary[bits - 1 - i] = ((num >> i) & 1) + '0';
```

This uses:

- Right shift `>>`
- Bitwise AND `&`

Shift and AND operations are very fast CPU-level operations.

# 7. Important Interview / Exam Answer

Question:

```text
Does using >> and & reduce the time complexity?
```

Answer:

```text
No, using >> and & does not reduce the Big-O time complexity.
Both division method and bitwise method have O(bits) time complexity because every bit must be generated.

However, the bitwise method is faster in practical execution because shift and AND operations are simpler and faster than division and modulus operations.
So bitwise method improves the constant factor, not the Big-O complexity.
```

# 8. Final Article

Decimal to binary conversion can be done in different ways in C.  
The common method uses division and modulus operators. In this method, the decimal number is repeatedly divided by 2, and the remainder is used as the binary digit. This is easy to understand and suitable for beginners.

Another method uses bitwise operators such as right shift `>>` and bitwise AND `&`. In this method, each bit of the number is extracted directly. The expression `(num >> i) & 1` gives the bit at position `i`.

Both methods require a loop that runs for the number of bits. Therefore, the time complexity of both methods is `O(bits)`. If 8 bits are required, the loop runs 8 times. If 32 bits are required, the loop runs 32 times.

The bitwise method does not reduce the Big-O time complexity, but it is faster in practical execution. This is because bitwise operations are simpler for the CPU compared to division and modulus operations.

So, the final conclusion is:

```text
Division method time complexity: O(bits)
Bitwise method time complexity: O(bits)
Bitwise method is faster in practice, but complexity remains the same.
```
